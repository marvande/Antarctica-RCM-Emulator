```python
from google.cloud import storage

def read_file_blob(bucket_name, destination_blob_name, destination_file_name):
    """Read a file from the bucket."""
 
    storage_client = storage.Client()
    bucket = storage_client.bucket(bucket_name)
    blob = bucket.blob(destination_blob_name)
 
    # read as string
    #read_output = blob.download_as_string()
    read_output = blob.download_to_filename(destination_file_name)
    
    path = "gs://ee-downscalingclimatemodels/year_ACCESS1-3.nc2"
    #df = pd.read_csv(path)
    #df = xr.open_dataset(path, decode_times=False)
    print(
        "File {} read successfully  from Bucket  {} to {}.".format(
            destination_blob_name, bucket_name, destination_file_name
        )
    )
    return read_output
```
